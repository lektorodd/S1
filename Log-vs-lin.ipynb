# Logaritmisk vs lineær andreakse

Ser på funksjonen $f(x)=2\cdot 1.01^x, D_f = [0,500]$

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f(x):
    return 2 * 1.01**x
```

```python
x = np.linspace(0,500,250)
y = f(x)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (10,5))
fig.suptitle('Lineær skala vs logaritmisk skala', fontsize = 18)
ax1.plot(x, y)
ax2.plot(x, y)
plt.yscale("log")
```
